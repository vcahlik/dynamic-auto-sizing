## Anytime grid search

```python
def _raw_relative_combinations_for_level(level, n_range_arguments):
    l = level + 1
    single_argument_values = tuple(i / (2 ** l) for i in range(1, 2 ** l))
    all_argument_values = [single_argument_values for _ in range(n_range_arguments)]
    return list(itertools.product(*all_argument_values))

def get_relative_combinations_for_level(level, n_range_arguments):
    relative_combinations = set(_raw_relative_combinations_for_level(level, n_range_arguments))
    previous_combinations = set()
    for i in range(0, level):
        previous_combinations = previous_combinations.union(set(_raw_relative_combinations_for_level(i, n_range_arguments)))
    relative_combinations = list(relative_combinations.difference(previous_combinations))
    random.shuffle(relative_combinations)
    return relative_combinations
```

```python
f = plt.figure()

points = get_relative_combinations_for_level(0, 2)
x = [x[0] for x in points]
y = [x[1] for x in points]
plt.scatter(x, y, c='black', marker='s', label="Level 1")

points = get_relative_combinations_for_level(1, 2)
x = [x[0] for x in points]
y = [x[1] for x in points]
plt.scatter(x, y, color=(0.25, 0.25, 0.25), marker='x', label="Level 2")

points = get_relative_combinations_for_level(2, 2)
x = [x[0] for x in points]
y = [x[1] for x in points]
plt.scatter(x, y, c='grey', marker='2', label="Level 3")

points = get_relative_combinations_for_level(3, 2)
x = [x[0] for x in points]
y = [x[1] for x in points]
plt.scatter(x, y, color=(0.85, 0.85, 0.85), marker='.', label="Level 4")

plt.xlabel(r'$a$', fontsize=13)
plt.ylabel(r'$b$', fontsize=13)
legend = plt.legend(fontsize=11)
frame = legend.get_frame()
frame.set_alpha(1)

plt.xlim(0, 1)
plt.ylim(0, 1)

f.savefig("anytime-grid-search.pdf", bbox_inches='tight')
```

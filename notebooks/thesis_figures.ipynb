```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use('seaborn-paper')

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
# set matplotlib font to Times
from matplotlib import rc
rc('font',**{'family':'serif','serif':['Times']})
rc('text', usetex=True)
```

## L1 vs. L2 regularization

```python
f = plt.figure()

x = np.linspace(-1.5, 1.5, num=200)
# plt.plot(x, (x**2), 'k', label=r"$\alpha=0$")
# plt.plot(x, (x**2) * ((np.sign(x) + 0.2) ** 2), 'k:', label=r"$\alpha=0.2$")
plt.plot(x, x ** 2, 'k--', label=r'$l_2$')
plt.plot(x, abs(x), 'k', label=r'$l_1$')
plt.xlabel(r'$\theta{}_i$', fontsize=13)
plt.ylabel(r'$\Omega(\theta{}_i)$', fontsize=13)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.legend(fontsize=11, loc='lower right')

plt.xlim(-1.5, 1.5)
plt.ylim(0, 2)

f.savefig("l1-12-regularization.pdf", bbox_inches='tight')
```

## Grid search

```python
import itertools

f = plt.figure()

x = np.linspace(1, 3, num=3)
y = np.linspace(1, 3, num=3)
product = list(itertools.product(x, y))
x = [x[0] for x in product]
y = [x[1] for x in product]

plt.scatter(x, y, c='black')
plt.vlines(x, 0, y, linestyle="dotted", colors='grey')
plt.hlines(y, 0, x, linestyle="dotted", colors='grey')
# plt.plot(x, abs(x), 'k', label=r'$l_1$')
plt.xlabel(r'$a$', fontsize=13)
plt.ylabel(r'$b$', fontsize=13)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
# plt.title('Grid search', fontsize=16)
# plt.legend(fontsize=11, loc='lower right')

plt.xlim(0.5, 3.5)
plt.ylim(0.5, 3.5)

f.savefig("grid-search.pdf", bbox_inches='tight')
```

## Random search

```python
f = plt.figure()

x = np.random.uniform(0.5, 3.5, 9)
y = np.random.uniform(0.5, 3.5, 9)

plt.scatter(x, y, c='black')
plt.vlines(x, 0, y, linestyle="dotted", colors='grey')
plt.hlines(y, 0, x, linestyle="dotted", colors='grey')
# plt.plot(x, abs(x), 'k', label=r'$l_1$')
plt.xlabel(r'$a$', fontsize=13)
plt.ylabel(r'$b$', fontsize=13)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
# plt.title('Random search', fontsize=16)
# plt.legend(fontsize=11, loc='lower right')

plt.xlim(0.5, 3.5)
plt.ylim(0.5, 3.5)

f.savefig("random-search.pdf", bbox_inches='tight')
```

# Figures for reference

## Activation functions

```python
f = plt.figure()

x_start = -3
x_end = 3

x = np.linspace(x_start, x_end)
relu = np.linspace(x_start, x_end)
elu = np.linspace(x_start, x_end)

relu[relu<0] = 0
relu[relu>0] = x[x>0]

elu[elu<0] = np.exp(x[x<0]) - 1
elu[elu>0] = x[x>0]

plt.plot(x, 1 / (1 + np.exp(-x)), 'k', label=r"Sigmoid")
plt.plot(x, relu, 'k:', label=r"ReLU")
plt.plot(x, elu, 'k-.', label=r"ELU")
plt.xlabel('$x$', fontsize=13)
plt.ylabel('$f(x)$', fontsize=13)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.legend(fontsize=11);

plt.grid(True)
plt.xlim(-3, 3)
plt.ylim(-1, 2)

f.savefig("activation-functions.pdf", bbox_inches='tight')
```
